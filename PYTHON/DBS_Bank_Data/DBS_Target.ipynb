```python
import pandas as pd
PD_DBS = pd.read_csv('PD 2023 Wk 1 Input.csv')
PD_target= pd.read_csv('Targets.csv')
PD_DBS_bck = PD_DBS.copy()
PD_target_bck = PD_target.copy()
PD_DBS_bck
```

```python
PD_target_bck
```

```python
PD_DBS_bck[['Bank Name','Transaction ID']] = PD_DBS_bck['Transaction Code'].str.split("-",1,expand=True)
PD_DBS_bck
```

```python
Replace_str = {1 : 'online' , 2 : 'In-person'}
PD_DBS_bck['Online or In-Person'] = PD_DBS_bck['Online or In-Person'].replace(Replace_str)
PD_DBS_bck
```

```python
PD_DBS_bck['Transaction Date'] = pd.to_datetime(PD_DBS_bck['Transaction Date'],format='%d/%m/%Y %H:%M:%S')
PD_DBS_bck
```

```python
PD_DBS_bck['Transaction Date'] = PD_DBS_bck["Transaction Date"].dt.quarter
```

```python
PD_DBS_bck
```

```python
Transaction_Value_by_Type = PD_DBS_bck.groupby(['Online or In-Person','Transaction Date'])['Value'].sum().reset_index()
Transaction_Value_by_Type.columns = ['Online or In-Person','Quarter', 'Value']
Transaction_Value_by_Type['Value'] = PD_DBS_bck.groupby(['Online or In-Person', 'Transaction Date'])['Value'].transform('sum')

```

```python
Transaction_Value_by_Type
```

```python
target_columnName = list(PD_target_bck.columns)
modified_columns = []
for col in target_columnName:
    modified_columns.append(col.replace('Q',''))
modified_columns
```

```python
PD_target_bck.columns=modified_columns
PD_target_bck
```

```python
PD_DBS_bck
```

```python
Transaction_Value_by_Type
```

```python
pivoted_targets = PD_target_bck.melt(id_vars='Online or In-Person', var_name='Quarter', value_name='quarterly targets')
pivoted_targets
```

```python

#Transaction_Value_by_Type['Quarter'] = Transaction_Value_by_Type['Quarter']
#Transaction_Value_by_Type = Transaction_Value_by_Type.astype('str')
Transaction_Value_by_Type['Quarter'] = Transaction_Value_by_Type['Quarter'].astype(str)
Transaction_Value_by_Type['Online or In-Person'] = Transaction_Value_by_Type['Online or In-Person'].str.title()
pivoted_targets['Online or In-Person'] = pivoted_targets['Online or In-Person'].str.title()

merged_data = pd.merge(Transaction_Value_by_Type, pivoted_targets, on=['Online or In-Person', 'Quarter'])
merged_data
```

```python
merged_data['Variance to Target'] = merged_data['Value'] - merged_data['quarterly targets']

merged_data
```

```python
merged_data.to_csv('output.csv', index=False)
```
