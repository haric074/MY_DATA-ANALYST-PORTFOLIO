```python
import pandas as pd
# Read CSV file
Match_summary = pd.read_csv('Match_summary.csv')
# Create a copy of the DataFrame
Match_summary_copy = Match_summary.copy()
# Split 'Result' column into 'Winner' and 'Margin'
Match_summary_copy[['Winner','Margin']] = Match_summary_copy.Result.str.split("won",1,expand=True)
Match_summary_copy.head(5)
```

#DROp Unnecessary Columns

```python
# Drop unnecessary columns
Drop_columns = ['Result','Margin']
Match_summary_copy = Match_summary_copy.drop(Drop_columns,axis=1)
```

```python
# Get unique teams
Teams_list = Match_summary_copy.Team1.unique()
```

```python
# Create initial Points Table
Points_Table = pd.DataFrame(Teams_list)
Points_Table['M'] = '9' 
Points_Table.columns='Country','M'
Points_Table = pd.DataFrame(Points_Table.sort_values(by='Country').reset_index(drop = True))
```

```python
# Calculate the number of wins for each team
winner_counts = Match_summary_copy.groupby('Winner')['Winner'].count().reset_index(name='Count')
winner_counts.columns = 'Country','Count'
Points_Table ['W'] = winner_counts['Count'].astype(int)
# Calculate the number of losses for each team
Points_Table['L'] = Points_Table['M'].astype(int) - Points_Table['W']
Points_Table
```

```python
# Save the Points Table to a CSV file
Points_Table.to_csv('Points_Table.csv')
```

Player Batting stats

```python
Player_B_stats = pd.read_csv('Batting_stats.csv')
Player_B_stats_copy = Player_B_stats.copy()
```

```python
Player_B_stats_copy['Boundy%'] = ((((Player_B_stats_copy['4s'] * 4) + (Player_B_stats_copy['6s'] * 6)) / Player_B_stats_copy['Runs']) * 100).round(2)
Player_B_stats_copy.head()
HighestRuns=Player_B_stats_copy.copy()
HighestRuns.to_csv('HighestRuns.csv')
```

```python
Power_Players = pd.DataFrame(Player_B_stats_copy.sort_values(by='Boundy%',ascending=False)).reset_index(drop = True)
Power_Players.head()
```

```python
Power_Players.to_csv('Players_with_Max_Boundry_%.csv')
```

Bowling_ECo

```python
Player_Bowling_Ec = pd.read_csv('Bowling_stats.csv')
Player_Bowling_Ec_copy = Player_Bowling_Ec.copy()
Player_Bowling_Ec_copy = Player_Bowling_Ec_copy.dropna()
Player_Bowling_Ec_copy.head()
```

```python
Player_Bowling_Ec_copy['Wkts'] = Player_Bowling_Ec_copy['Wkts'].astype(int)
Player_Bowling_Ec_copy['Runs'] = Player_Bowling_Ec_copy['Runs'].astype(int)
Player_Bowling_Ec_copy['ER'] = Player_Bowling_Ec_copy.Runs / Player_Bowling_Ec_copy.Overs
Player_Bowling_Ec_copy = Player_Bowling_Ec_copy.sort_values(by ='ER').reset_index(drop = True).round(2)
Player_Bowling_Ec_copy.head()
```

```python
Player_Bowling_Ec_copy.to_csv('Player_Bowling_ER.csv')
```

Most_Wickets

```python
Most_Wikcet = pd.DataFrame(Player_Bowling_Ec_copy.sort_values(by = 'Wkts',ascending = False)).reset_index(drop = True)
Most_Wikcet['Wkts'] = Most_Wikcet['Wkts'].astype(int)
Most_Wikcet.head()
```
