```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
inputdata = pd.read_csv('Cars_Data1.csv')
inputdata.head()
```

```python
inputdata_copy = inputdata.copy()
inputdata_copy.head()
```

```python
inputdata_copy.shape
```

```python
inputdata_copy.isnull().sum()
```

```python
#Dropping isnull row in CSV
inputdata_copy = inputdata_copy.dropna(how = 'all')
inputdata_copy.isnull().sum()
```

```python
#Filling isnan values
inputdata_copy.Cylinders.fillna(inputdata_copy.Cylinders.mean(),inplace = True)
inputdata_copy.isnull().sum()
```

```python
plt.figure(figsize=(5,7))
inputdata_copy['Make'].value_counts().plot(kind='bar', color='navy')
plt.title('Number of Cars by Make')
plt.xlabel('Make')
plt.ylabel('Number of Cars')
plt.show()
```

```python
inputdata_copy.Origin.unique()
```

```python
plt.figure(figsize=(5,7))
inputdata_copy['Origin'].value_counts().plot(kind='bar', color='navy')
plt.title('Number of Cars by Origin')
plt.xlabel('Origin')
plt.ylabel('Number of Cars')
plt.show()
```

```python
inputdata_copy.Invoice = inputdata_copy.Invoice.str.replace('$','')
inputdata_copy.head()
```

```python
inputdata_copy.shape
```

```python
inputdata_copy.Invoice = inputdata_copy.Invoice.str.replace(',','')
inputdata_copy.Invoice = inputdata_copy.Invoice.astype(int)
print(inputdata_copy['Invoice'].dtype)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate the average price for each make
average_price_by_make = inputdata_copy.groupby('Make')['Invoice'].mean()

# Select the top 5 makes with the highest average prices
top_5_makes = average_price_by_make.nlargest(5).index

# Filter the data for the top 5 makes
top_5_data = inputdata_copy[inputdata_copy['Make'].isin(top_5_makes)]

# Create a bar plot for the top 5 makes
plt.figure(figsize=(5, 8))
sns.barplot(x='Make', y='Invoice', data=top_5_data, hue='Origin', palette='muted')

# Set labels and title
plt.title('Top 5 Car Makes by Average Price', fontsize=20)
plt.xlabel('Make', fontsize=10)
plt.ylabel('Invoice', fontsize=10)
plt.legend(loc='upper left', fontsize=14)
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```

```python
sns.set_style("whitegrid")
# create a bar plot with larger size
plt.figure(figsize=(5,7))
sns.barplot(x = 'Make', y = 'Invoice', data = inputdata_copy.head(50), hue = 'Origin', palette='Blues')

#set labels and titles
plt.title("Price Distribution by No. of Make", fontsize=16)
plt.xlabel("Make",fontsize=8)
plt.ylabel("Price",fontsize=8)
```

```python
#groupBy_cylinder = inputdata_copy.groupby('Make')['Cylinders'].mean()
car_data_grouped = inputdata_copy.groupby('Make').mean()[['Cylinders']]
car_data_grouped.sort_values(by ='Cylinders',ascending = False,inplace = True)
top_10_car_types = car_data_grouped.head(10)
top_10_car_types
```

```python
plt.figure(figsize=(5,7))
#create a bar plot of the average cylinder size by top 10 car types
sns.barplot(x=top_10_car_types.index, y='Cylinders', palette='PuBu', data=top_10_car_types)

#set the title and labels
plt.title("Average Cylinder Size by Top 10 Car Types", fontsize=16)
plt.xlabel("Car Type")
plt.ylabel("Average Cylinder Size")
plt.xticks(rotation=45) # to rotate x_axis label by 45 degrees
plt.show()
```

```python
#Plot size
plt.figure(figsize=(14,8))

#create a scatter plot of engine size vs horsepower
sns.scatterplot(x='EngineSize', y='Horsepower', data=inputdata_copy, hue='Origin', palette ='Set1')

#set the title, labels and size of the title
plt.title("Relation between Engine Size and Horsepower by Origin", fontsize=15)
plt.xlabel("Engine Size", fontsize=12)
plt.ylabel("Horsepower", fontsize=12)
plt.legend(loc='upper left') # to set location of legend
plt.show()
```

```python
#plot size
plt.figure(figsize=(20,10))

sns.lmplot(x='EngineSize', y='Horsepower', data=inputdata_copy, hue='Origin',palette = 'YlOrBr', fit_reg=True)
#set the title and labels
plt.title("Relation between EngineSize and Horsepower by Origin", fontsize=16)
plt.xlabel("Engine Size")
plt.ylabel("Horsepower")
plt.show()
```
